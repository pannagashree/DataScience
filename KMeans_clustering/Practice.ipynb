**K-means Clustering**

Used to cluster unlabeled data in unsupervised learning

Choosing K value:    
One method is **elbow method**    
Compute the sum of squared error(SSE) for some values of K    
SSE is the sum of squared distance between each member of cluster and it's centroid.    
SSE decreases as K gets larger.    
The idea of elbow is that to choose the K at whcih SSE decreases abruptly(producing an elbow effect in graphs)

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs #for creating artificial datasets
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
len(data)
```

```python
data[0].shape  #data[0] is a numpy array
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
data[1]  #it is the cluster which it belongs to
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
#kmeans=KMeans(n_clusters=8)
#kmeans=KMeans(n_clusters=3)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ #predicted clusters
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6)) #sharey means same axes for both the subplots
#first argument in subplots is number of rows and 2nd argument is number of columns
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
